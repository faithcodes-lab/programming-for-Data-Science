## Q1.1: Reading in rows of a CSV file. [10 marks]

Write a function `csv_rows(filename, from_row=None, to_row=None)`
to read in the header and specific rows of a CSV file.
The input arguments of the function should be a string specifying a filename,
a starting row number
`from_row` and
a final row number `to_row`.
The row numbers should be either integers or None.
The starting row number can be 1 or higher or None.
Row 1 refers to the first row after the header.
If the starting row number is None, then the default value of 1 should be used.
The ending row number can be 1 or higher or None.
The ending row number refers to the last row to be included.
If the ending row number is None, then all rows starting at the starting row should be read in.
The function should return two outputs:
a list with the column names and list of rows, where each row is a list of values (strings or numbers).

You can use the file `part1.csv` for your tests.

The function below reads the header and selected rows from the csv file provided by the user. It allows the user to specify a starting row (from_row) and an ending row (to_row), where row 1 refers to the first row after the header. If the user does not provide a starting or ending row, the function treats from_row and to_row as None and applies the default behaviour.

This fuction below returns a list of column names extracted from the header of the csv file and list of rows, where each row is a list of values selected based on the starting and ending row.  

```python
#
# Code can go here.
import csv

def csv_rows(filename, from_row=None, to_row=None):
    
    """
    This function reads a csv file and returns the header and selected rows by the user.

    parameters:
              filename (string): which is the CSV file path
              from_row (Integer): Starting row(indexed 1) and can be  optional(default to 1) 
              to_row (Integer): Ending row which can be optional(Default to none) that is end of the file.
     
     Returns:
            tuple: A list of the column names and data rows
    """

    # Default starting row as per assignment requiremnet
    if from_row is None:
        from_row = 1

    #header stores the first row of the csv file
    header = None

    #Lists which stores the filtered data rows
    filtered_rows = []

    #count the filtered data rows excluding the headr
    rows_count = 0   

    # read the csv file
    with open(filename, newline="", encoding="utf-8") as file:
        for row in csv.reader(file):

            #store the header row
            if header is None:
                header = row
                continue

            #increase the row count
            rows_count += 1

            #skip rows before the starting row
            if rows_count < from_row:
                continue

            #stop reading if we reach the ending row
            if to_row is not None and rows_count > to_row:
                break

            #append the row to the filtered row list
            filtered_rows.append(row)

    return header, filtered_rows
```

## Q1.2: Reading in Columns of a CSV file. [10 marks]

Write a function `csv_cols(filename, from_row=None, to_row=None, columns=None)`
that takes as inputs a filename, a starting row, a final row and a list of column names.
The same conventions as for Q1.1 should be used for the starting and final row input parameters.
The column name parameter can either be a list of column names or None, which indicates all columns.
The function should return a dictionary with the selected column names as
keys and a list of values (strings or numbers) for the corresponding column.
Your function should use the function from Q1.1.

You can use the file `part1.csv` for your tests.



This function extracts selected columns from a csv file and returns a dictionary of column names and values, re using the row selection function in Q1.1. if no columns are specified, the default behaviour is to return all columns.

```python
def csv_cols(filename, from_row=None, to_row=None, columns=None):
    """
    This function reads a csv file and returns selected columns and rows by the user.

    parameters:
            filename (string): which is the csv file path.
            from_row (Int, opional): Starting row((1-indexed, after the header). Defaults to 1 if not provided.
            to_row (Int, optional): Ending row, If None, rows are read until the end of the file.
            columns (list, optional): list of column names extracted, if None, all columns are returned.
    Returns:
            dict: A dictionary where Keys are column names and the values are list of the data for each selected columns.
    """
    
    #Reuse Q 1.1 function as specified by assignment instruction
    header, rows = csv_rows(filename, from_row, to_row)

    # Default reading, if no columns are specified, use all columns

    if columns is None:
        columns = header

    # store column data in a dictionary
    column_data = {}

    #loop through the seleceted columns
    for col in columns:
        if col in header:
            col_index = header.index(col)
            #retreive values from each row
            column_data[col] = [row[col_index] for row in rows]
        else:
            raise ValueError("Column not found in the csv header")
    
    return column_data
```

## Q1.3: Calculating the Euclidean Distance between two lists of values [10 marks]

Write a function `euclidean_distance(p, q)` to calculate
the Euclidean Distance between two lists of values.
The two input lists should be of equal length.
The function should return the distance value.



This function compute the Euclidean distance between two list of numerical values. The two input list must be of equal length, otherwise, the function raises a valueError

```python
import math

def euclidean_distance(p, q):
    """
    This fuction computes the Euclidean distance between two points p and q.
    
    parameters: 
    p (list): First list of numerical values 
    q(list) : Second list of numerical values

    returns:
    A float: The euclidean distance between points p and q assuming both have equal lenght.

    """

    if len(p)!= len(q):
        raise ValueError("Input lists must have the same length.")

    #calculate the squared differneces
    squared_differences = [(x - y) ** 2 for x, y in zip(p, q)]
    
    #sum the squared differences
    sum_squared_differences = sum(squared_differences)
    
    #take the square root of the sum to get the Euclidean distance
    distance = math.sqrt(sum_squared_differences)
    
    return distance
```

## Q1.4: Calculating a Table of Pair-Wise Distances [10 marks]

Write a function `distance_table(data, columns=None)` that takes as inputs
a dictionary with column names as keys and a list of values for each key.
The `columns` parameter can be either a list of column names or None, which
means all column names should be used.
The function should return the list of column names used and
a table of pair-wise distances between the columns, represented as a list of lists.
Your function must use the function from Q1.3.

You can use the file `part1.csv` and the function from Q1.2 for your tests.

This function calculates the Pair-Wise distance between the selected columns of the data set. The distances are computed by reusing the euclidean distance function in Q1.3

```python
def distance_table(data, columns=None):
    
    """
    Computes a table of pair-wise distances between selected columns in the data set.

    Parameters:
        data (dict): Dictionary with column names as keys and lists of values.
        columns (list, optional): List of column names. If None, all columns are used.

    Returns:
        tuple:
             list of column names used
             lists representing the distance table
    """
    # Use all columns if none is specified

    if columns is None:
        columns = list(data.keys())
        
    # Empty list that stores the distance table
    table = []

    #compute the pair-wise distance
    for p in columns: #p represent the first column
        row = []
        for q in columns: #q represent the second column

            #convert the string to float
            col1 = [float(x) for x in data[p]]
            col2 = [float(x) for x in data[q]]
            distance = euclidean_distance(col1, col2)
            row.append(distance)
        table.append(row)

    return columns, table  
```

## Q1.5: Develop a Function to Print a Custom Table [10 marks]

Write a function `print_distance_table(columns, table, ...)`
that prints the distance table in a clear and attractive way.
The input parameters should be a list of column names
and a table of distances as returned by the function from Q1.4.
You can add extra parameters for further customisation.

You can use the file `part1.csv` and the functions from Q1.2 and Q1.4 for your tests.


The function below prints the pair-wise distance table generated in q1.4 in a clear and readable format.
It displays the column names as the headers and the values of the rounded distance for each pair of columns.

```python
def print_distance_table(columns, table, column_width=12):

    """
    This function prints a pair-wise dustance table in a clear and readable format.

    Parameters:
         columns(list): List of column names.
         table(list): Pair-wise distanace table.
         column_width(int, optional): The size of the width of each column
    """
    #print header row
    header_row = "".join(f"{column:>{column_width}}" for column in [""] + columns)
    print(header_row)
    print("-" * len(header_row))

    #print distance rows
    for column_name, row in zip(columns, table):
        row_str = "".join(f"{round(value, 4):>{col_width}}" for value in row)
        print(f"{col_name:>{col_width}}{row_str}")
```
