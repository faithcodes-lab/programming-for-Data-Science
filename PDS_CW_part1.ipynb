## Q1.1: Reading in rows of a CSV file. [10 marks]

Write a function `csv_rows(filename, from_row=None, to_row=None)`
to read in the header and specific rows of a CSV file.
The input arguments of the function should be a string specifying a filename,
a starting row number
`from_row` and
a final row number `to_row`.
The row numbers should be either integers or None.
The starting row number can be 1 or higher or None.
Row 1 refers to the first row after the header.
If the starting row number is None, then the default value of 1 should be used.
The ending row number can be 1 or higher or None.
The ending row number refers to the last row to be included.
If the ending row number is None, then all rows starting at the starting row should be read in.
The function should return two outputs:
a list with the column names and list of rows, where each row is a list of values (strings or numbers).

You can use the file `part1.csv` for your tests.

The function below reads the header and selected rows from the csv file provided by the user. It allows the user to specify a starting row (from_row) and an ending row (to_row), where row 1 refers to the first row after the header. If the user does not provide a starting or ending row, the function treats from_row and to_row as None and applies the default behaviour.

This fuction below returns a list of column names extracted from the header of the csv file and list of rows, where each row is a list of values selected based on the starting and ending row.  

```python
#
# Code can go here.
import csv

def csv_rows(filename, from_row=None, to_row=None):
    
    """
    This function reads a csv file and returns the header and selected rows by the user.

    parameters:
              filename (string): which is the CSV file path
              from_row (Integer): Starting row(indexed 1) and can be  optional(default to 1) 
              to_row (Integer): Ending row which can be optional(Default to none) that is end of the file.
     
     Returns:
            tuple: A list of the column names and data rows
    """

    # Default starting row as per assignment requiremnet
    if from_row is None:
        from_row = 1

    #header stores the first row of the csv file
    header = None

    #Lists which stores the filtered data rows
    filtered_rows = []

    #count the filtered data rows excluding the headr
    rows_count = 0   

    # read the csv file
    with open(filename, newline="", encoding="utf-8") as file:
        for row in csv.reader(file):

            #store the header row
            if header is None:
                header = row
                continue

            #increase the row count
            rows_count += 1

            #skip rows before the starting row
            if rows_count < from_row:
                continue

            #stop reading if we reach the ending row
            if to_row is not None and rows_count > to_row:
                break

            #append the row to the filtered row list
            filtered_rows.append(row)

    return header, filtered_rows
```

## Test code 1: Default values (from_row = None, to_row= None)

```python
#Test 1: Deafult readings which returns header and all rows
header, rows = csv_rows("part1.csv", from_row=None, to_row=None)

print("Header: ")
print(header)

#proof that the funtion printed the whole rows    
print("\nNumber of data rows returned:")
print(len(rows))

#print just first five lines for readability
print("\nFirst 5 lines:")
for row in rows[:5]:
    print(row) 
```
